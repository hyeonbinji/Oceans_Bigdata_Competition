```python
# datetime 형태로 변환
def int_to_datetime(x):
    return datetime.strptime(str(x), '%Y%m%d%H%M%S')

data['BERTHN_DTM'] = data['BERTHN_DTM'].apply(lambda x: int_to_datetime(x))
```

```python
data['DEP_DTM'] = pd.to_datetime(data['DEP_DTM'], format='%Y%m%d%H%M%S', errors='coerce')
```

```python
data['REG_DATE'] = data['REG_DATE'].apply(lambda x: int_to_datetime(x))
```
