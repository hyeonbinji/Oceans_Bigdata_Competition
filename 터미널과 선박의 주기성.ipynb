```python
terminals = list(data['TERMINAL'].unique())
ships = list(data['VESSEL_CD'].unique())
```

```python
from tqdm.notebook import tqdm
```

```python
#10분정도 걸림
cycle = {}

for i, terminal in tqdm(enumerate(terminals)):
    for j, ship in enumerate(ships):
        dummy = smalldata[smalldata['TERMINAL'] == terminal]
        dummy = dummy[dummy['VESSEL_CD'] == ship]
        
        
        
        if len(dummy) > 1:
            key = (terminal, ship)
            counts = len(dummy)
            time_diff = []
            dummy['BERTHN_DTM'] = pd.to_datetime(dummy.loc[:,'BERTHN_DTM'])
            for k in range(counts-1):
                diff1 = (dummy.iloc[k+1,2] - dummy.iloc[k,2]).seconds / 60
                diff2 = (dummy.iloc[k+1,2] - dummy.iloc[k,2]).days * 1440
                time_diff.append(np.abs(diff1 + diff2))

            cycle[key] = [time_diff, np.mean(time_diff), counts]
            
            
            
            
cycle            
```

```python
cycle_df = pd.DataFrame(cycle, index = ['시간 차', '주기(시간 차 평균)', '방문 횟수']).T
cycle_df
```

```python
cycle_df.to_csv('터미널과 선박의 주기성.csv')
```
