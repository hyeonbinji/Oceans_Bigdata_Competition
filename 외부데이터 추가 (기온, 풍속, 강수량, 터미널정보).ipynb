```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
```

```python
wea_2019 = pd.read_csv('2019년 기상.csv', encoding = 'cp949')
wea_2020 = pd.read_csv('2020년 기상.csv', encoding = 'cp949')
wea_2021 = pd.read_csv('2021년 기상.csv', encoding = 'cp949')
wea_2022 = pd.read_csv('2022년 기상.csv', encoding = 'cp949')
weather = pd.concat([wea_2019, wea_2020, wea_2021, wea_2022], ignore_index = True)
data = pd.read_csv('2.2. 트레드링스 data_통합.csv', low_memory = False)
```

```python
weather = weather.fillna(0)
```

```python
def int_to_datetime(x):
    return datetime.strptime(str(x)[:10], '%Y%m%d%H')

data['일시'] = data['BERTHN_DTM'].apply(lambda x: int_to_datetime(x))
data
```

```python
busan = ['BPTC','HKTL','BIT','PNC','HJNC','PNIT','DPCT','HPNT','BNMT','BNCT','DDCT','BCT']
kwangyang = ['GWCT','KITL','SMGT']
incheon = ['HJIT','SNCT','ICT','IFPC','E1CT']
pyeongtaek = ['PCTC','PNCT']
ulsan = ['UNCT', 'JUCT']
kunsan = ['IGCT']
location = []

for i in data['TERMINAL']:
    if i in busan:
        location.append('부산')
    elif i in kwangyang:
        location.append('광양시')
    elif i in incheon:
        location.append('인천')
    elif i in pyeongtaek:
        location.append('수원')
    elif i in ulsan:
        location.append('울산')
    else:
        location.append('군산')
        
data['지점명'] = location
data
```

```python
str(data['일시'][1])[:16]
```

```python
def time_to_str(x):
    return str(x)[:16]

data['일시'] = data['일시'].apply(lambda x: time_to_str(x))
data
```

```python
data1 = pd.merge(data, weather, on = ['지점명', '일시'], how = 'left')
data1.head()
```
